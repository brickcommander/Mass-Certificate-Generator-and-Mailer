```python
#####################FILES####################
excel_file = 'data.xlsx'
certi_template = 'template5.jpg'
###################################################



#####################################################################################################################
################################################DRAG THING AND POSITIONS########################################
####################################################################################################################
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
rects = []
img = cv2.imread(certi_template,1)

#img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
default = img.copy()
original = img.copy()
#variables

#true while mouse button down and false while mouse button up
drawing = False
ix,iy = -1,-1


#fn
def draw_rectangle(event,x,y,flags,params) :
    global ix,iy,drawing,rects,img,default
    if event==cv2.EVENT_LBUTTONDOWN :
        drawing = True
        ix,iy = x,y
        
    elif event==cv2.EVENT_MOUSEMOVE:
        if drawing==True:
            img = default.copy()
            cv2.imshow('drag',img)
            cv2.rectangle(img,(ix,iy),(x,y),(0,0,0),2)

            
    elif event==cv2.EVENT_LBUTTONUP:
        if drawing == True :
            drawing = False
            img = default.copy()
            cv2.imshow('drag',img)
           
            cv2.rectangle(img,(ix,iy),(x,y),(0,0,0),2)
            cv2.imshow('drag',img)
            if (cv2.waitKey(3000) & 0xFF  ==110) : #n
                cv2.imshow('drag',img)
                print("Ignored")
                img = default.copy()
                cv2.imshow('drag',img)
            else :
                cv2.imshow('drag',img)
                rects.append([(ix,iy) , (x,y)])
                print("Appended")
                default = img.copy()
            cv2.imshow('drag',img)

img = default.copy()
cv2.imshow('drag' , img)
cv2.setMouseCallback('drag', draw_rectangle)


while True :
    cv2.imshow('drag' , img)
    if cv2.waitKey(1) & 0xFF ==27 :
        break
    
cv2.destroyAllWindows()
default = original.copy()
img = default.copy()





#######################################################################
##################EXCEL READING AND PRINTING#####################
##################################################################

xlsx = pd.ExcelFile(excel_file)
certi_sheets = []
font = cv2.FONT_HERSHEY_SIMPLEX
for sheet in xlsx.sheet_names:
    certi_sheets.append(xlsx.parse(sheet))
for certi_sheet in certi_sheets :
    for position in certi_sheet :
        for position in range(certi_sheet.shape[0]) :
            text_box = 0
            img = original.copy()
            for label in certi_sheet.columns: 
                pos = (rects[text_box][0][0],rects[text_box][1][1])
                text_box+=1
                cv2.putText(img , text = str(certi_sheet.at[position,label]), org = pos, fontFace=font,fontScale=1,color=(0,0,0),thickness=1,lineType=cv2.LINE_AA)
            cv2.imwrite('Generated/'+str(position)+'.png',img) 
```
